```python
def push_user_input(text):
  # Create the chat and poll the status
  chat_poll = coze.chat.create_and_poll(
      bot_id=bot_id,
      user_id=user_id,
      additional_messages=[
          Message.build_user_question_text(text),
      ],
  )

  # Print the response messages from the chat
  for message in chat_poll.messages:
      print(message.content, end="", flush=True)

  # Check if the chat is completed and print token usage
  if chat_poll.chat.status == ChatStatus.COMPLETED:
      print()
      print("Token usage:", chat_poll.chat.usage.token_count)
```
